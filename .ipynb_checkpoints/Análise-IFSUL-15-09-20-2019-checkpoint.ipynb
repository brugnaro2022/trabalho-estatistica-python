```python
colunas_cursos = [
    'Instituição',
    'Unidade de Ensino',
    'Tipo de Curso',
    'Código da Matricula',
    'Tipo Oferta'
]

df_cursos = df[colunas_cursos]
df_cursos.head()
```

```python
df[df['Instituição'] == 'IFSUL']['Unidade de Ensino'].drop_duplicates().count()
```

```python
# Verificando quantidade de matrículas
df['dt_ocorrencia_matricula'] = pd.to_datetime(df['Data de Ocorrencia da Matricula'], format='%d/%m/%Y')
df['ano_ocorrencia_matricula'] = df['dt_ocorrencia_matricula'].dt.year
df['dt_fim_ciclo'] = pd.to_datetime(df['Data de Fim Previsto do Ciclo'], format='%d/%m/%Y')
df['ano_fim_ciclo'] = df['dt_fim_ciclo'].dt.year

df_ifsul_2019 = df[(df['Instituição'] == 'IFSUL')]
df_ifsul_2019.shape
```

```python
f = df_ifsul_2019[
    (df_ifsul_2019['Tipo Oferta'] != '-')
    & (df_ifsul_2019['Tipo Oferta'] != 'Concomitante')
    & (df_ifsul_2019['Tipo Oferta'] != 'Subsequente')
]

f.groupby(['Tipo Oferta','Renda Familiar'])['Renda Familiar'].count()
```

```python
f = df_ifsul_2019[
    (df_ifsul_2019['Tipo Oferta'] != '-')
    & (df_ifsul_2019['Tipo Oferta'] != 'Concomitante')
    & (df_ifsul_2019['Tipo Oferta'] != 'Subsequente')
]

a = f[f['Situação de Matrícula'] == 'Concluída']
b = a.groupby(['Tipo Oferta','Cor / Raça'])['Cor / Raça'].count()
b
```

```python
# Filtrando tipo de oferta por situação
f = df_ifsul_2019[
    (df_ifsul_2019['Tipo Oferta'] != '-')
    & (df_ifsul_2019['Tipo Oferta'] != 'Concomitante')
    & (df_ifsul_2019['Tipo Oferta'] != 'Subsequente')
]

f.groupby(['Tipo Oferta', 'Situação de Matrícula'])['Tipo Oferta'].count()
```
